## Задание 1

Переведите содержимое файла purchase_log.txt в словарь purchases вида:
```python
{'1840e0b9d4': 'Продукты', ...}
```

````python
import json
````

````python
i = 0
purchases_keys = []
purchases_values = []
with open('purchase_log.txt', 'r', encoding = 'utf-8') as f:
#     next(f)
    for line in f:
        line = line.strip()        
        dict_i = json.loads(line)

        purchases_keys.append(dict_i['user_id'])
        purchases_values.append(dict_i['category'])

purchases = dict(zip(purchases_keys, purchases_values))
````

## Задание 2

Для каждого user_id в файле visit_log.csv добавьте третий столбец с категорией покупки (если покупка была).

````python
# здесь не придумал как корректно произвести запись в тот же файл

with open('visit_log_new.csv', 'w+') as fw:

    with open('visit_log.csv') as vis:
        for ln in vis:
            i = ln.strip()
            
            if purchases.get(i[:10]) == None:
                fw.write(ln)
            else: fw.write('{},{}\n'.format(i, purchases.get(i[:10])))
````

## Задание 3

Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории

````python
with open('funnel.csv', 'w+') as f_write:
    with open('visit_log_new.csv', 'r') as f_read:
        for ln in f_read:
            if ln.count(',') == 2:
                f_write.write(ln)
````
