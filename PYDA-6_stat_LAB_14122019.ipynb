# Гипотезы:


* wagon класс автомобилей с наибольшей колесной базой



* convertible и hardtop — наиболее дорогие классы автомобилей

```python
convertible = data[data['body-style']=='convertible']
hatchback = data[data['body-style']=='hatchback']
sedan = data[data['body-style']=='sedan']
wagon = data[data['body-style']=='wagon']
hardtop = data[data['body-style']=='hardtop']
```

```python
# wagon.info()
```

```python
from scipy.stats import ttest_ind
ttest_ind(data['wheel-base'], wagon['wheel-base'])
```

**Смотрим на графике:**

```python
w_wheel_b = wagon['wheel-base']
wheel_b = data['wheel-base']
sns.distplot(w_wheel_b)
sns.distplot(wheel_b)
plt.title('Плотность вероятности распределения wheel_b')
plt.legend(['wagon', 'all car'])
plt.show()
```

```python
print(ttest_ind(data['price'], convertible['price']))
c_price = convertible['price']
all_prices = data['price']
sns.distplot(c_price)
sns.distplot(all_prices)
plt.title('Плотность вероятности распределения price')
plt.legend(['convertible', 'all car'])
plt.show()
```

```python
print(ttest_ind(data['price'], hardtop['price']))
h_price = hardtop['price']
sns.distplot(h_price)
sns.distplot(all_prices)
plt.title('Плотность вероятности распределения price')
plt.legend(['hardtop', 'all car'])
plt.show()
```
