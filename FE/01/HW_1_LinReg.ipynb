```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

#### Беру набор данных с [сайта](https://1xbet.whoscored.com/Matches/1415830/LiveStatistics/Italy-Serie-A-2019-2020-Lazio-Udinese)

```python
df = pd.read_excel('foot_stat.xlsx', header=1, sheet_name=0)
```

```python
df
```

```python
# from IPython.display import HTML
# HTML('<img src="">')
```

#### Мы знаем, что у нас отсутствует один столбец с [данными](https://prnt.sc/qho9cb), который сильно влияет на `Rating`. Смотрим как влияют текущие данные.

```python
from sklearn.metrics import mean_absolute_error

X = df[['Shots', 'ShotsOT','KeyPasses','PA%','AerialsWon','Touches']]
y = df['Rating']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(X)
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
print('\n',reg.score(X,y))
```

#### Оставляю 4 признака с наибольшим весом

```python
X = df[['Shots', 'ShotsOT', 'KeyPasses', 'AerialsWon']]
y = df['Rating']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(df[['Shots', 'ShotsOT', 'KeyPasses', 'AerialsWon']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
print('\n',reg.score(X,y))
```

#### Введу относительный признак точности удара по воротам ` t = ShotsOT / Shots `

```python
df['t'] = df['ShotsOT'] / df['Shots']
df = df.fillna(0)
df
```

### Посмотрим на предсказание с новым признаком:

```python
X = df[['KeyPasses', 'AerialsWon', 't']]
y = df['Rating']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(df[['KeyPasses', 'AerialsWon','t']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
print('\n',reg.score(X,y))
```

#### Добавлю удары `Shots`:

```python
X = df[['Shots','KeyPasses', 'AerialsWon', 't']]
y = df['Rating']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(df[['Shots','KeyPasses', 'AerialsWon','t']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
print('\n',reg.score(X,y))
```
