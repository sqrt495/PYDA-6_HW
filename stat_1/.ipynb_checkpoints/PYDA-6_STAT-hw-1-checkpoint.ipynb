### Вопрос: О чем говорит знак коэффициента?

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
Freedom = df['Freedom']
norm_Freedom = Freedom - statistics.mean(Freedom)
print(statistics.mean(norm_Freedom))
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

sizing_Freedom = min_max_scaler.fit_transform(np.array(norm_Freedom).reshape(-1, 1))
print(min(sizing_Freedom))
print(max(sizing_Freedom))
```

Написать временные предсказания цен на авокадо типа organic для выбранного региона: LasVegas, начиная с августа 2017 года до конца марта 2018. В качестве тренировочных данных использовать данные до июля 2017 года включительно.
В качестве примера использовать лекционный материал в lecture_1.ipynb

```python
# подготовим датасет:

df['Date'] = pd.to_datetime(df['Date'])

PREDICTION_TYPE = 'organic'
df_conv = df[df.type == PREDICTION_TYPE]

regions = df_conv.groupby(df_conv.region)

PREDICTING_FOR = "LasVegas"
date_price = regions.get_group(PREDICTING_FOR)[['Date', 'AveragePrice']].reset_index(drop=True)

rcParams['figure.figsize'] = 15,6
date_price.plot(x='Date', y='AveragePrice', kind="line")
```

```python
# пытался сделать по-своему, получился такой ужас, буду потом разбираться:

date_price_2 = df[ (df['region'] == PREDICTING_FOR) ][['Date', 'AveragePrice']].reset_index(drop=True)


rcParams['figure.figsize'] = 15,6
date_price_2.plot(x='Date', y='AveragePrice', kind="line")
```

```python
# подготовим данные для модели:

date_price = date_price.rename(columns={'Date':'ds', 'AveragePrice':'y'})

date_price = date_price.sort_values(by='ds')

amount_80 = int(len(date_price)*0.80)
train_price_80 = date_price[:amount_80]
test_price_80 = date_price[amount_80:]
# 80% - тренировачные данные до июля
train_price_80.tail()
```

```python
test_price_80.tail()
```

```python
from fbprophet import Prophet
m = Prophet()
m.fit(train_price_80)

# 240 - как раз до конца марта, все как в лекции:
future = m.make_future_dataframe(periods=240)
forecast = m.predict(future)
fig1 = m.plot(forecast)
```

```python
plt.figure(figsize=(15,4))
forecast_future = forecast.iloc[135:,:]

fig, axs = plt.subplots(2)
fig.suptitle('Сравниваем. плавный - прогноз, резкий - реальность:')
axs[0].plot(forecast_future['ds'], forecast_future['yhat'])
axs[1].plot(test_price_80['ds'], test_price_80['y'])

plt.show()
```

```python
# вместе:
plt.figure(figsize=(15,4))
forecast_future = forecast.iloc[135:,:]
plt.plot(forecast_future['ds'], forecast_future['yhat'])
plt.plot(test_price_80['ds'], test_price_80['y'])
plt.show()
```
