### 5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными

**Гипотеза №1:**



Сотрудники `management` работают в компании в среднем дольше остального персонала, значит они больше дургих удовлетворены работой. Сравню с `hr`.

```python
management_sat=df[df["sales"]=="management"]["satisfaction_level"]
hr_sat=df[df["sales"]=="hr"]["satisfaction_level"]
```

```python
len(management_sat), len(hr_sat)
```

```python
hr_sat=hr_sat.sample(630)
statistics.variance(management_sat), statistics.variance(hr_sat)
```

```python
fig = plt.figure()

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.hist(management_sat)
ax_2.hist(hr_sat)

ax_1.set(title = 'management_sat', xticks=[], yticks=[])
ax_2.set(title = 'hr_sat', xticks=[], yticks=[])

plt.show()
```

Дисперсия схожая, но распределение нельза назвать нормальным. Тем не менее:

```python
ttest_ind(management_sat, hr_sat, equal_var=False)
```

**Вывод:**


Существенной разницы в данных нет, вероятность ошибки весьма мала. Значит `management` и `hr` одинаково ценят свою работу, не смотря на сроки службы в компании.

**Гипотеза №2:**



Как мы знаем из графика выше сотрудники с наибольшим `average_montly_hours` имеют много проектов и слабо удовлетворены работой, полагаем, что у них более высокий показатель `Work_accident`. Сравню персонал с большим и средним кол-вом проектов:

```python
df.number_project.describe()
```

```python

many_projects = df[df['number_project'].isin([6,7])]["Work_accident"]
mean_projects=df[df["number_project"]==4]["Work_accident"]
```

```python
# сравниваем
len(many_projects), len(mean_projects)
```

```python
mean_projects=mean_projects.sample(1430)
print(many_projects.shape, mean_projects.shape)
```

**Тк распределение биноминальное, проверим просто cравнив кол-во инцендентов в выборках:**

```python
 print(f'Количество инцендентов у сотрдников с 6-7 проектами = {many_projects.eq(1).values.sum()}.\nКоличество инцендентов у сотрдников с 4 проектами = {mean_projects.eq(1).values.sum()}.\nДанные в расчете на 1430 человек.')
```

**Вывод:** количество проектов не влияет на неприятные случаи на работе.

```python
print('\n',ttest_ind(many_projects, mean_projects),'\n')

sns.distplot(many_projects)
sns.distplot(mean_projects)
plt.title('Плотность распределения Work_accident')
plt.legend(['many_projects', 'mean_projects'])
plt.show()
```

### 8. Показать распределение сотрудников по зарплатам совместно с департаментами

Выведу данные по всем сотрудникам за историю компании, тех что не работают более и тех кто работает до сих пор.

```python
total_s = pd.get_dummies(df['salary'], prefix = 'salary_')
```

```python
total_s['deps'] = df.sales
total_s['left'] = df.left
```

```python
working_now = total_s.query('left == 0').groupby('deps').sum()
del working_now['left']

no_working = total_s.query('left == 1').groupby('deps').sum()
del no_working['left']

total_workers = total_s.groupby('deps').sum()
del total_workers['left']
```

```python
dep_list= list(total_workers.index)
dep_list.insert(0, 'x')
dep_list
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(nrows = 3, ncols =1)

x = range(0,10)

   
ax[0].set_title('total_workers', fontsize=20, loc = 'left')
ax[1].set_title('working_now', fontsize=20, loc = 'left')
ax[2].set_title('no_working', fontsize=20, loc = 'left')
    
def make_hist(_, df):  
    ax[_].bar(x, df.salary__low, width = 0.6, color='grey', label='low salary')
    ax[_].bar(x, df.salary__medium, width = 0.4, color='navy', label='medium salary')
    ax[_].bar(x, df.salary__high, width = 0.2, color='lime', label='high salary')
    ax[_].xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax[_].set_xticklabels(dep_list)
    ax[_].legend(loc = 'upper left')
    ax[_].grid()

make_hist(0, total_workers)  
make_hist(1, working_now) 
make_hist(2, no_working) 

fig.set_figwidth(16)    
fig.set_figheight(16)   

plt.show()
```

### 10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
no_working['total_L_by_dep'] = no_working.sum(axis=1)

working_now['total_N_by_dep'] = working_now.sum(axis=1)


summary = pd.DataFrame({'working_now' : working_now['total_N_by_dep'],
                       'no_working' : no_working['total_L_by_dep']},
                       index=dep_list[1:])
```

```python
summary['work%'] = summary['working_now']/14999
summary['work%'] = summary['work%'].map('{:,.2f}'.format)
summary['left%'] = summary['no_working']/14999
summary['left%'] = summary['left%'].map('{:,.2f}'.format)
```
