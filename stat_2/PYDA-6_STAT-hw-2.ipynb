# Задание:
Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
# проверяю отсутствующие значения
data.info()
# отделю показатели от ДФ, преабразую в массивы
h_coast = np.array(data['MEDV']).reshape(-1,1)
low_stat = np.array(data['LSTAT']).reshape(-1,1)
```

```python
# импорт
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()

# разбиваем выборку
X_train, X_test, y_train, y_test = train_test_split(
    h_coast,
    low_stat,
    test_size = 0.2,
    random_state = 10
)

# учим
model.fit(X_train, y_train)
```

```python
print('Вероятность ошибки прогноза: {:.2%}'.format((1 - model.score(X_test, y_test))))
```

```python
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 8,5


plt.xlabel('Цена жилья')
plt.ylabel('Процент жителей с низким уровнем жизни')
plt.title('Линейная регрессия:')

# прямая
x = np.linspace(-10, 60, 20)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)

# точки
plt.plot(h_coast, low_stat, 'r.')
plt.xlim(-10,80)
plt.ylim(-10,60)
plt.show()
```

# Задание:
Добавить новую колонку в таблицу - 'continent'. Каждая страна относится к определенному материку. Всего материко 6:
*   Евразия,

*   Австралия,
*   Антарктида.

Таким образом, мы можем классифицировать каждый элемент как член определенного класса - континента. 

```python
# scrapping wiki

import re
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('https://simple.wikipedia.org/wiki/List_of_countries_by_continents')
bs = BeautifulSoup(html, "html.parser")
```

```python
continentList = bs.findAll('span', {'class': 'toctext'})
continent_list = []
for continent in continentList:
    continent_list.append(continent.get_text())
```

```python
continent_list =  continent_list[:-1]
continent_list.remove('Antarctica')
continent_list
```

```python
# take all <ul>-tags
uls = bs.find_all(['ul'])
uls_content = []
for i in uls:
    uls_content.append(re.split('[\n]', i.get_text()))
```

```python
# страны хранятся в списке со столицами:
countries_with_capitals = uls_content[1:7]
```

```python
#оставим только названия стран в списках

countries = []
cc_ = []
for el in countries_with_capitals:
    for i in el:
        try:
            cc_.append(re.findall('^[\w]*', i)[0].replace(' ', ''))
            cc_.append(re.findall('^[\w\s]*', i)[0].replace(' ', ''))
            cc_.append(re.findall('^[\w\s\(\)ô\']*(?= -)', i)[0].replace(' ', ''))
        except: pass
    
    countries.append(set(cc_))
    cc_ *=0 
# print(countries)
    
# создадим словарь:

def make_country_list(continent_list, countries):
    globus = {}
    for a in range(0,6):
         globus[continent_list[a]]=countries[a]
    return globus

globus = make_country_list(continent_list, countries)
```

```python
globus
```

```python
def continent_checker(row):
    for c_ in list(globus.keys()):
        if row['Country (region)'].replace(' ', '') in (globus[c_]):
            return c_
    
data['continent'] = data.apply(continent_checker, axis=1)
data.info()
```
